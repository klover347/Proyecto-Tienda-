```python
Paso 1: Cargar las Bibliotecas Necesarias

Primero, importa las bibliotecas de Python necesarias para trabajar con los datos y crear gráficos.

# Importar bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
Paso 2: Cargar los Datos CSV

Asegúrate de que los archivos CSV de las tiendas estén subidos a Google Colab o Google Drive. Aquí está el código para cargar los datos desde un archivo CSV.

# Cargar los archivos CSV de cada tienda (ajusta la ruta del archivo según donde se encuentre)
store1 = pd.read_csv('/path/to/store1.csv')
store2 = pd.read_csv('/path/to/store2.csv')
store3 = pd.read_csv('/path/to/store3.csv')
store4 = pd.read_csv('/path/to/store4.csv')

# Verificar que los datos se cargaron correctamente
print(store1.head())
print(store2.head())
print(store3.head())
print(store4.head())
Paso 3: Análisis de los Datos
1. Facturación Total por Tienda
# Calcular la facturación total de cada tienda
facturacion_store1 = store1['Ventas'].sum()
facturacion_store2 = store2['Ventas'].sum()
facturacion_store3 = store3['Ventas'].sum()
facturacion_store4 = store4['Ventas'].sum()

print(f'Tienda 1 - Facturación Total: {facturacion_store1}')
print(f'Tienda 2 - Facturación Total: {facturacion_store2}')
print(f'Tienda 3 - Facturación Total: {facturacion_store3}')
print(f'Tienda 4 - Facturación Total: {facturacion_store4}')
2. Categorías Más Vendidas por Tienda
# Agrupar por tienda y categoría, luego sumar ventas
categorias_mas_vendidas_store1 = store1.groupby('Categoría')['Ventas'].sum()
categorias_mas_vendidas_store2 = store2.groupby('Categoría')['Ventas'].sum()
categorias_mas_vendidas_store3 = store3.groupby('Categoría')['Ventas'].sum()
categorias_mas_vendidas_store4 = store4.groupby('Categoría')['Ventas'].sum()

print("Categorías más vendidas Tienda 1: ")
print(categorias_mas_vendidas_store1)
3. Promedio de Evaluación por Tienda
# Agrupar por tienda y calcular la calificación promedio
promedio_calificacion_store1 = store1['Calificación'].mean()
promedio_calificacion_store2 = store2['Calificación'].mean()
promedio_calificacion_store3 = store3['Calificación'].mean()
promedio_calificacion_store4 = store4['Calificación'].mean()

print(f'Promedio de calificación Tienda 1: {promedio_calificacion_store1}')
print(f'Promedio de calificación Tienda 2: {promedio_calificacion_store2}')
print(f'Promedio de calificación Tienda 3: {promedio_calificacion_store3}')
print(f'Promedio de calificación Tienda 4: {promedio_calificacion_store4}')
4. Productos Más y Menos Vendidos
# Productos más vendidos por tienda
productos_vendidos_store1 = store1.groupby('Producto')['Ventas'].sum()
producto_mas_vendido_store1 = productos_vendidos_store1.idxmax()
producto_menos_vendido_store1 = productos_vendidos_store1.idxmin()

print(f"Producto más vendido Tienda 1: {producto_mas_vendido_store1}")
print(f"Producto menos vendido Tienda 1: {producto_menos_vendido_store1}")
5. Costo Promedio de Envío por Tienda
# Calcular el costo promedio de envío
costo_envio_promedio_store1 = store1['Costo_envio'].mean()
costo_envio_promedio_store2 = store2['Costo_envio'].mean()
costo_envio_promedio_store3 = store3['Costo_envio'].mean()
costo_envio_promedio_store4 = store4['Costo_envio'].mean()

print(f"Costo promedio de envío Tienda 1: {costo_envio_promedio_store1}")
print(f"Costo promedio de envío Tienda 2: {costo_envio_promedio_store2}")
print(f"Costo promedio de envío Tienda 3: {costo_envio_promedio_store3}")
print(f"Costo promedio de envío Tienda 4: {costo_envio_promedio_store4}")
Paso 4: Crear Gráficos de Visualización
1. Gráfico de barras para comparar las ventas totales por tienda
# Crear gráfico de barras para las ventas totales de cada tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturacion = [facturacion_store1, facturacion_store2, facturacion_store3, facturacion_store4]

plt.bar(tiendas, facturacion, color='skyblue')
plt.title('Ventas Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ventas Totales')
plt.show()
2. Gráfico circular para mostrar la distribución de categorías más vendidas en Tienda 1
categorias_mas_vendidas_store1.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Distribución de Categorías más Vendidas - Tienda 1')
plt.ylabel('')
plt.show()
3. Gráfico de dispersión para comparar ventas y costos de envío en Tienda 1
plt.scatter(store1['Ventas'], store1['Costo_envio'], color='orange')
plt.title('Relación entre Ventas y Costo de Envío - Tienda 1')
plt.xlabel('Ventas')
plt.ylabel('Costo de Envío')
plt.show()
```
